## **Randomized Search CV**

### Logistic regression

```python
parameters1 = {'penalty':('l1','l2'), 
              'C':(1, 10, 100, 1000)}
```

```python
clf = RandomizedSearchCV(LogisticRegression(), parameters1, random_state=0)
clf.fit(X_train, y_train)
print(clf.best_params_)
print(f"Tarda {clf.cv_results_['mean_fit_time'].mean()} per training")
```

### RandomForestClassifier

```python
parameters = {
    'n_estimators':(10,50,100,200),
    'criterion': ('gini', 'entropy'),
    'min_samples_split': (2,5,10)
}
```

```python
clf = RandomizedSearchCV(RandomForestClassifier(), parameters, random_state=0)
clf.fit(X_train, y_train)
print(clf.best_params_)
print(f"Tarda {clf.cv_results_['mean_fit_time'].mean()} per training")
```
